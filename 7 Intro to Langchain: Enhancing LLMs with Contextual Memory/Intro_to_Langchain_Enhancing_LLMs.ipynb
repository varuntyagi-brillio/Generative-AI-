### Prompt template

A PromptTemplate in LangChain is a powerful tool that helps in creating dynamic prompts for large language models (LLMs). It allows you to define a template with placeholders for variables that can be filled in with actual values at runtime. This helps in managing and reusing prompts efficiently, ensuring consistency and reducing the likelihood of errors in prompt creation.

A PromptTemplate consists of:

- **Template String**: The actual prompt text with placeholders for variables.
- **Input Variables**: A list of variables that will be replaced in the template string at runtime.

there are other ways of passing data to variables like f string etc, but the usage of prompt templates shines as we keep adding more things to the setup

````python
from langchain import LLMChain, PromptTemplate


# Define the prompt template
product_description_template = PromptTemplate(
    input_variables=["product_name", "product_features"],
    template="""
    You are a creative copywriter. Write an engaging product description for the following product:

    Product Name: {product_name}
    Features: {product_features}

    The description should be catchy and highlight the unique selling points of the product.
    """
)
````

````python
# Define a function to generate the product description using the template
def generate_product_description(product_name, product_features):
    # Initialize the LLMChain with the prompt template
    chain = LLMChain(llm=llm, prompt=product_description_template)
`
    # Run the chain with the provided product details
    description = chain.invoke({
        "product_name": product_name,
        "product_features": product_features
    })

    return description
````

here we have created a LLM chain, which we are going to have a look below

````python
# Example usage
product_name = "Smart Home Assistant"
product_features = "Voice control, AI-powered, Energy efficient, Multi-room support"
description = generate_product_description(product_name, product_features)

print("Generated Product Description:")
print(description)
````

### Chains in Langchain

In LangChain, a "chain" refers to a sequence of steps where each step involves processing some input and producing some output, typically using a large language model (LLM) or other tools. Chains allow for creating complex workflows by linking together multiple prompts or actions. This helps in building more sophisticated and structured applications, enhancing modularity, and reusability.


**Chains can consist of:**

- Sequential Steps: Each step depends on the output of the previous step.
- Parallel Steps: Steps that can be executed independently.

Chains provide a structured way to manage interactions with LLMs and other tools, making it easier to handle complex logic and stateful interactions.

**Extra:**

In LangChain, both PromptTemplate and ChatPromptTemplate serve as tools to structure and manage prompts for large language models (LLMs). However, they are used for different purposes and in different contexts. Here's a comparison of the two:

### 1. PromptTemplate
- Purpose:

PromptTemplate is used to create structured, single-turn prompts for LLMs. It’s ideal for scenarios where you want to generate text based on a specific input or set of inputs without maintaining any context between turns.
- Structure:

It typically consists of a template string with placeholders for variables that can be filled in at runtime. It’s designed for generating one-shot or single-response outputs from the LLM.
- Use Case:

Single-turn tasks like generating product descriptions, summarizing text, answering factual questions, etc.

### 2. ChatPromptTemplate
- Purpose:

ChatPromptTemplate is designed for multi-turn conversations or chat-based interactions with LLMs, especially with models like GPT-4 that support a chat-based API. It allows you to manage a sequence of messages, including user inputs, assistant responses, and potentially system instructions.
- Structure:

It allows for the creation of a sequence of messages that simulate a conversation. Each message can be from a user, an assistant, or the system, and the template can contain placeholders for dynamic content.
- Use Case:

Multi-turn conversational agents, interactive assistants, chatbots, or any scenario where context needs to be maintained across multiple exchanges.

````python
from langchain.chains import SequentialChain, LLMChain
from langchain.prompts import ChatPromptTemplate

# prompt template 1: translate to english
first_prompt = ChatPromptTemplate.from_template(

    "Translate the following review to english:"
    "\n\n{Review}"
)
# chain 1: input= Review and output= English_Review
chain_one = LLMChain(llm=llm, prompt=first_prompt,
                     output_key="English_Review"
                    )
````

so to create a chain we basically combine an llm and a prompt.

- also notic we passed on output key, which is actually the key to the output in the final output dictionary for that part of the chain.
- so the output key for the first chain is the english review

````python
second_prompt = ChatPromptTemplate.from_template(
    "Can you summarize the following review in 1 sentence:"
    "\n\n{English_Review}"
)
# chain 2: input= English_Review and output= summary
chain_two = LLMChain(llm=llm, prompt=second_prompt,
                     output_key="summary"
                    )
````

- notice how we passed the english review as an input to the next part of the chain

````python
# prompt template 3: translate to english
third_prompt = ChatPromptTemplate.from_template(
    "What language is the following review, answer with just the name of the language:\n\n{Review}"
)
# chain 3: input= Review and output= language
chain_three = LLMChain(llm=llm, prompt=third_prompt,
                       output_key="language"
                      )
````

- here when we check the original language we pass tht as the review
- so now you understand how to mix n match inputs and outputs

````python
# Chain 4: Analyze the sentiment of the summary
fourth_prompt = ChatPromptTemplate.from_template(
    "Analyze the sentiment of the following summary and output it in any of these: (positive, negative, neutral):\n\nSummary: {summary}"
)
chain_four = LLMChain(llm=llm, prompt=fourth_prompt, output_key="sentiment")
````

here in the output we are taking all of the outputs back, but if we want we can choose what we need as a output, it can be any one or all or couple of chains

````python
# Create the overall SequentialChain
overall_chain = SequentialChain(
    chains=[chain_one, chain_two, chain_three, chain_four],
    input_variables=["Review"],
    output_variables=["English_Review", "summary", "language", "sentiment"],
    verbose=True
)

# Run the SequentialChain
review = 'एसर का नया एडवांस आई-सीरीज 43-इंच टीवी देखने में जितना आकर्षक है, इस्तेमाल करने में उतना ही स्मूथ। इसकी पिक्चर क्वालिटी और साउंड दोनों ही शानदार हैं। स्मार्ट फीचर्स के मामले में भी ये टीवी पीछे नहीं है, गूगल असिस्टेंट सपोर्ट के साथ आप आराम से वॉइस कमांड दे सकते हैं। हालांकि, हैंड्स-फ्री फीचर की कमी थोड़ी खलती है। इसके अलावा, टीवी के साथ दिया गया रिमोट भी काफी अच्छा है, सभी जरूरी बटन और OTT प्लेटफॉर्म के लिए डेडिकेटेड बटन भी मौजूद हैं। कुल मिलाकर, अगर आप एक अच्छा और किफायती स्मार्ट टीवी ढूंढ रहे हैं तो एसर का ये टीवी आपके लिए एक बढ़िया विकल्प हो सकता है।'
result = overall_chain({"Review": review})
print(result)
````

### Router chains

In LangChain, Router Chains are used to dynamically select and route the input to different sub-chains or components based on the input's characteristics or conditions. This is useful when you need to handle different types of inputs or tasks with specialized processing logic.

Example Scenario
Let's consider an example where you have a system that processes customer feedback. Depending on the content of the feedback, you might want to route it to different chains:

- Positive Feedback: Summarize and then generate a thank-you response.
- Negative Feedback: Summarize and then generate an apology response with a follow-up question for more details.
- Neutral Feedback: Summarize and then generate a generic response.

image.png

-  notice the similarity to classification of input we did before in llm apis lecture

````python
from langchain.chains.router import MultiPromptChain
from langchain.chains.router.llm_router import LLMRouterChain,RouterOutputParser
from langchain.prompts import PromptTemplate
````

````python
positive_feedback_template = """You are an empathetic customer service representative.
Summarize the following positive feedback and generate a thank-you response.
Here is the feedback:
{input}"""

negative_feedback_template = """You are an empathetic customer service representative.
Summarize the following negative feedback and generate an apology response with a follow-up question for more details.
Here is the feedback:
{input}"""

neutral_feedback_template = """You are an empathetic customer service representative.
Summarize the following neutral feedback and generate a generic response.
Here is the feedback:
{input}"""
````

````python

prompt_infos = [
    {
        "name": "positive",
        "description": "Good for processing positive feedback",
        "prompt_template": positive_feedback_template
    },
    {
        "name": "negative",
        "description": "Good for processing negative feedback",
        "prompt_template": negative_feedback_template
    },
    {
        "name": "neutral",
        "description": "Good for processing neutral feedback",
        "prompt_template": neutral_feedback_template
    }
]
````

prompt_infos: This list stores dictionaries, each containing information about a specific type of feedback template. The keys in each dictionary are:

- "name": Identifier for the type of feedback.
- "description": A brief description of what the prompt is suited for.
- "prompt_template": The actual prompt template defined above.

````python
# Destination Chains Setup: A loop iterates over each item in prompt_infos,
# creates a chain for each feedback type,
# and stores it in the destination_chains dictionary.
# Each chain is created by combining the ChatPromptTemplate with LLMChain.


destination_chains = {}
for p_info in prompt_infos:
    name = p_info["name"]
    prompt_template = p_info["prompt_template"]
    prompt = ChatPromptTemplate.from_template(template=prompt_template)
    chain = LLMChain(llm=llm, prompt=prompt)
    destination_chains[name] = chain # this is what we'll pass to the router chain
````

````python
destination_chains
````

destination_chains: This dictionary will store chains corresponding to each feedback type, which will later be used by the router chain to determine the appropriate response pathway.

````python

destinations = [f"{p['name']}: {p['description']}" for p in prompt_infos] # this is what we'll pass to the template
destinations_str = "\n".join(destinations)
````

- Destinations String: This list comprehension and subsequent join create a string representation of all the possible destinations (types of feedback). This is needed for the router template.

This destinations_str will be used in the router prompt to inform the model about the different feedback processing options available.


````python
destinations
````

````python
destinations_str
````

````python

default_prompt = ChatPromptTemplate.from_template("reply to this {input} gracefully")


default_chain = LLMChain(llm=llm, prompt=default_prompt)
# This chain acts as a fallback when the router is unable to categorize the input. It generates a generic response.

MULTI_PROMPT_ROUTER_TEMPLATE = """Given a raw text input to a \
language model select the model prompt best suited for the input. \
You will be given the names of the available prompts and a \
description of what the prompt is best suited for. \

<< FORMATTING >>
Return a markdown code snippet with a JSON object formatted to look like:
```json
{{{{
    "destination": string \ name of the prompt to use or "DEFAULT"
    "next_inputs": string \ original input
}}}}
```

<< CANDIDATE PROMPTS >>
{destinations}

<< INPUT >>
{{input}}

<< OUTPUT (remember to include the ```json)>>"""
````

Router Template: This template instructs the model on how to choose the appropriate feedback chain based on the input. It describes the available prompt options and asks the model to select the most suitable one.

````python
router_template = MULTI_PROMPT_ROUTER_TEMPLATE.format(destinations=destinations_str)

router_prompt = PromptTemplate(
template=router_template,
input_variables=["input"],
output_parser=RouterOutputParser(),
)

router_chain = LLMRouterChain.from_llm(llm, router_prompt)
````

The router_chain is created by combining the router prompt with the LLM. It will determine which feedback type chain to use based on the user's input.
- we have not yet combined the chains, this router chain that we are creating now is to route to the next step or choose the next step

````python
chain = MultiPromptChain(router_chain=router_chain,
                         destination_chains=destination_chains,
                         default_chain=default_chain, verbose=True
                        )
````


### Explanation:

1. **Prompt Templates**:
   - Defined specific templates for positive, negative, and neutral feedback. Each template asks for a summary and generates an appropriate response.

2. **Destination Chains**:
   - Created `LLMChain` instances for each type of feedback using the respective templates and stored them in the `destination_chains` dictionary.


3. **Router Template**:
   - Defined a router template that determines which chain to use based on the input feedback.

4. **Router Chain**:
   - Used `LLMRouterChain` to create the router chain from the router prompt.

5. **MultiPromptChain**:
   - Combined the router chain, destination chains, and default chain into a `MultiPromptChain`.



This code routes customer feedback to appropriate chains and generates tailored responses based on the feedback content.

## Memory

The concept of memory in long-chain models refers to the ability of these models to retain information across multiple interactions, enhancing their ability to provide coherent and contextually relevant responses over time. This is in contrast to traditional use of language models (LLMs) through APIs, where each interaction is typically stateless, meaning the model doesn't remember previous interactions unless explicitly provided with past context.

**Concept of Memory in Long-Chain Models**
- Stateful Interactions: Long-chain models maintain a memory of past interactions. This memory can include user preferences, prior questions and answers, ongoing tasks, and any relevant context shared in previous sessions.

- Context Retention: These models can recall past interactions to provide more relevant and accurate responses. This includes remembering names, preferences, ongoing projects, and any specific details that have been shared.

- Dynamic Updating: The memory is updated dynamically as new interactions occur. This means the model can learn and adapt based on new information, ensuring that the responses are always based on the latest context.

### ConversationBufferMemory

This memory type stores the entire conversation history. Every interaction is saved sequentially in a buffer, allowing the model to access all past exchanges.


Ideal for scenarios where having the complete history is crucial, such as complex customer service interactions or long-term projects where all details might be relevant.

- Provides a complete context, ensuring that no information is lost.
- Simplifies the task of referencing any part of the conversation.

### ConversationBufferWindowMemory

This type of memory also stores the conversation history but within a fixed-size window. Only the most recent exchanges are retained in the buffer.


Useful for interactions where only the recent context is relevant, such as live support chats where earlier parts of the conversation become less important as it progresses.


- Manages memory usage more efficiently by limiting the buffer size.
- Keeps recent context readily available without overwhelming the system with too much data.

````python
from langchain.memory import ConversationBufferWindowMemory

memory = ConversationBufferWindowMemory(k=1)
````

````python
memory.save_context({"input": "Hi"},
                    {"output": "What's up"})

memory.save_context({"input": "Not much, just hanging"},
                    {"output": "Cool"})
````

### ConversationTokenBufferMemory

This memory type stores the conversation based on token count rather than the number of exchanges. It retains the conversation history up to a certain number of tokens.

Effective in managing memory usage for interactions where the length of the text (in tokens) is more critical than the number of exchanges, such as applications with strict token limits.

- Controls memory usage by focusing on token count, which is often a direct cost factor in API usage.
- Ensures that the conversation history stays within manageable limits for performance reasons.

### ConversationSummaryMemory

This memory type creates and maintains a summarized version of the conversation. Instead of storing all details, it retains key points and significant information, updating the summary as the conversation progresses.

Best suited for lengthy interactions where retaining a full history is impractical, but key points need to be remembered, such as in ongoing advisory or coaching sessions.

- Keeps memory usage low by summarizing rather than storing the full text.
- Ensures that important context is retained without overwhelming the system.

````python
from langchain.memory import ConversationSummaryBufferMemory
````
